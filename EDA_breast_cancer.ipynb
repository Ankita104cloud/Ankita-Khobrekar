# **1. Load dataset**

```python
!pip install pandas
import pandas as pd
data = pd.read_csv("/content/breast-cancer.csv")
```

# **2. Understanding dataset**

```python
data.shape
data.head()
data.columns
data.dtypes
data.info
```

# **3. Data quality checks**

```python
data.isnull().sum()
data.duplicated().sum()
data.describe
```

# **4. Univariate analysis**

```python
# categorical variable
import seaborn as sns
sns.countplot(data = data, x = "class")
```

```python
# Numerical variable kde: kernel density estimation
sns.histplot (data["age"], kde = True)
```

# **5.Bivariate Analysis**

```python
sns.countplot(data = data, x = 'tumor-size', hue = 'class')
```

```python
# to see if malignancy grade affects recurrence
sns.countplot(data = data, x = 'deg-malig', hue = 'class')
```

# **6. Correlation analysis**

```python
# selection od numerical columns and correlation matrix and form heat map
# to see which variables are strongly associated
numeric_data = data.select_dtypes(include = 'number')
corr = numeric_data.corr()
sns.heatmap(corr, annot = True, cmap = 'coolwarm')
```

# **7. Feature vs Target analysis**

```python
# determine which variables are important
import seaborn as sns
sns.countplot (data = data, x = "age", hue = "class")
sns.countplot (data = data, x = 'breast-quad', hue = 'class')
sns.countplot (data = data, x = 'irradiat', hue = 'class')
```

# **8. Outliers detection (Numerical features)**

```python
#to check for extreme values
sns.boxplot(x = data['age'])
```
